```python
from lomond import WebSocket
import json
websocket = WebSocket('wss://ws-feed.gdax.com')
data=[]
for event in websocket:
    if event.name == "ready":
        websocket.send_json(
            type='subscribe',
            product_ids=['BTC-USD'],
            channels=['ticker']
        )
    elif event.name == "text":
        print(event.json)
        x=event.json
        data.append(x)
        with open('data.json', 'w') as outfile:
            json.dump(data, outfile)
            
            
```
