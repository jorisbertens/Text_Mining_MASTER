```python
from tweepy import Stream
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import json
import sentiment_mod as s
from textblob import TextBlob
from project_functions import *
```

```python
ckey = ""
csecret = ""
atoken = ""
asecret = ""

class listener(StreamListener):
    
    def on_data(self, data):
        all_data = json.loads(data)
        
        keys = ["created_at", "text"]
        tweet_dict = {key:value for (key,value) in all_data.items() if key in keys}
        
        tweet_dict = filter_english(tweet_dict)
        tweet_dict = removal_function(tweet_dict)
        
        
        tweet = tweet_dict["text"]
        sentiment_value, confidence = s.sentiment(tweet)     
        print(tweet, sentiment_value, confidence)
        
        if confidence*100 >=70:
            output = open("twitter-out.txt", "a")
            output.write(sentiment_value)
            output.write("\n")
            output.close()
            
        return True
    
    def on_error(self, status):
        print(status)
        
auth = OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)
twitterStream = Stream(auth, listener())
twitterStream.filter(track=["bitcoin"])
```
